###### Exploratory Data Analysis

```python
df.head()
```

```python
cat_var = []
con_var = []

for column in df.columns:
    if df[column].nunique()<=25 and df[column].dtype == 'object': # If column has less that 25 unique values then declaring it categorical
        cat_var.append(column)
    else:
        con_var.append(column)   # if column has > 25 unique values then declare it continuous
        
```

```python
for var in cat_var:
    plt.figure(figsize=(10,6))
    sns.set_style("darkgrid")
    plt.title("Count Of Type Of {}".format(var))
    ax = sns.countplot(df[var])
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
    ax.set_ylabel("Frequency Of %s" %var)
    
    for p in ax.patches:
        ax.annotate("{:.1f}".format(p.get_height()),(p.get_x(),p.get_height()))
        
    plt.show()
```

## Modelling

###### Spitting Datasets

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Item_Outlet_Sales'],inplace=False)
Y= df['Item_Outlet_Sales']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8)
```

###### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
dt = DecisionTreeRegressor(max_depth=15)
```

```python
dt.fit(X_train,Y_train)
```

```python
dt.score(X_train,Y_train)
```

```python
dt.score(X_test,Y_test)
```

### RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor().fit(X_train,Y_train)
```

```python
rfr.score(X_train,Y_train)
```

```python
rfr.score(X_test,Y_test)
```

#### Hyper-Prameter Tuning For Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(10,150,15)]

max_depth = [int(x) for x in np.linspace(5,40,8)]

min_samples_split = [int(x) for x in np.linspace(100,1500,15)]

max_features = ['auto',"sqrt"]
```

```python
params = {
    
    "n_estimators":n_estimators,
    "max_depth":max_depth,
    "min_samples_split":min_samples_split,
    "max_features":max_features
}
```

```python
tuner = RandomizedSearchCV(estimator=rfr,param_distributions=params,cv=5,n_jobs=1,verbose=2,n_iter=10)
```

```python
tuner.fit(X,Y)
```

```python
tuner.best_params_
```

```python
del rfr
```

```python
rfr = RandomForestRegressor(n_estimators=90,min_samples_split=200,max_features='auto',max_depth=15)
```

```python
rfr.fit(X_train,Y_train)
```

```python
rfr.score(X_train,Y_train)
```

```python
rfr.score(X_test,Y_test)
```

```python
Y_pred = rfr.predict(X_test)
```
